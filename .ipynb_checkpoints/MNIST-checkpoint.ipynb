## Prepare data

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

data = open('mnist_train.csv')
csv_file = csv.reader(data)

data_points = []

for row in csv_file:
    data_points.append(row)

# On enlève la première ligne, soit les "headers" de nos colonnes
data_points.pop(0)

data.close()

# converts strings to integer
data_points = np.int_(data_points)

# separates labels and inputs
y_train = data_points[:,0]
x_train = data_points[:,1:]
x_train_naive = np.round(x_train/255)

print(y_train)
print(y_train.shape)
print(x_train.shape, 'Chacune des 6000 images a 784 dimensions (pixels)')
```

```python
#prepare TEST data
data = open('mnist_test.csv')

csv_file = csv.reader(data)

data_points = []

for row in csv_file:
    data_points.append(row)

#On enlève la première ligne, soit les "headers" de nos colonnes
data_points.pop(0)


data.close()
data_points = np.int_(data_points)

y_test = data_points[:,0]
x_test = data_points[:,1:]

print(x_test.shape)

x_test_naive = np.round(x_test/255)

# create smaller TEST set
n_samples = 1000
idx = np.int_(10000 * np.random.rand(n_samples))

x_small_test = x_test[idx]
x_small_test_naive = x_test_naive[idx]
y_small_test = y_test[idx]
```
