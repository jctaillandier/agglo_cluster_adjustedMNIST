```python
%matplotlib inline
import csv
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from scipy.spatial.distance import squareform
import itertools
import time
```

### training set

```python
data = open('mnist_train.csv')

csv_file = csv.reader(data)

training_points = []

for row in csv_file:
    training_points.append(row)

# Remove "headers" of our csv file
training_points.pop(0)

data.close()

# Convert string into integer
training_points= np.int_(training_points)

# Pick a ration for splitting the digits into a training and a validation set.
ratio = int(len(training_points)*0.20)
validation_set = training_points[:ratio]
training_set = training_points[ratio:]

# Separates the first column, which are the labels/target from or input points
# for the validation and training set
y_validation = validation_set[:,0]
x_validation = validation_set[:,1:]
y_train = training_set[:,0]
x_train = training_set[:,1:]

# For algorithm implementation purpose, create smaller training set
small_y = y_train[:1000]
small_x = x_train[:1000]
small_set = training_set[:1000]
```

```python
# Make values 1,0
small_x_01 = np.zeros((small_x.shape[0],small_x.shape[1]))
for i in range(small_x.shape[0]):
    for j in range(small_x.shape[1]):
        if small_x[i,j] > 128:
            small_x_01[i,j] = 1
        else:
            small_x_01[i,j] = 0
            
```

### test set

```python
data = open('mnist_test.csv')

csv_file = csv.reader(data)

testing_points = []

for row in csv_file:
    testing_points.append(row)

# Remove "headers" of our csv file
testing_points.pop(0)


data.close()

# Convert string into integer
testing_points= np.int_(testing_points)
        
# Separates the first column, which are the labels (y_test)
# from or input points (x_test)
y_test = testing_points[:,0]
x_test = testing_points[:,1:]

```

```python
from scipy.ndimage import rotate
def RotateDigit(x, theta):
    x = np.reshape(x, (28,28))
    x = rotate(x, theta, axes=(1,0), reshape=False, output=None, order=3, mode='constant', cval=0.0, prefilter=True)
    #x = np.reshape(x, (748))
    return x
```

# Algorithms

### K-means without Sklearn

```python
def Kmeans(train_data, k, epsilon=0, dist='euclidean'):
    # List to store past centroid for graphing purposes
    past_centroids = []
    
    # Set the distance calculation type
    if dist == 'euclidean':
        dist_method = euclidean
        
    # Set the dataset
    data = train_data
    
    # Get the number of instances x (rows) and the number of feature (cols)
    num_x, num_features = data.shape
    
    # Choose k random centroids and set them to our list
    current_centroids = data[np.random.randint(0, num_x - 1, size=k)]
    old_centroids = np.zeros(current_centroids.shape)
    
    # For plotting purposes, keep track of previous centroids
    past_centroids.append(current_centroids)
    
    # Initialize variables to store the clusters
    clusters = np.zeros((num_x, 1))
    norm = dist_method(current_centroids, old_centroids)
    
    while norm > epsilon:
        norm = dist_method(current_centroids, old_centroids)
        old_centroids = current_centroids
        
        # For each instance in the dataset, define a distance vector of size k
        for index_x, x in enumerate(data):
            dist_vec = np.zeros((k, 1))
            
            # For each centroid, compute the distance between x and the centroid
            for index_centroid, centroid in enumerate(current_centroids):
                dist_vec[index_centroid] = dist_method(centroid, x)
            
            # Find the smallest distance, assign that distance to a cluster
            clusters[index_x, 0] = np.argmin(dist_vec)
            
        # Create a temporary list to keep track of the new centroids
        temp_centroids = np.zeros((k, num_features))
        
        # For each cluster, get all the points assigned to it and compute the mean
        for index in range(len(current_centroids)):
            points = [i for i in range(len(clusters)) if clusters[i] == index]
            centroid = np.mean(data[points], axis=0)
            temp_centroids[index, :] = centroid
            
        # Set the new centroids to the current list
        current_centroids = temp_centroids
        
        # Add our calculated centroids to our list of past centroids
        past_centroids.append(temp_centroids)
        
    # Return final calculated centrois, list of past centroids, and 
    # assignment for which cluster each data point belongs to
    return clusters, current_centroids

def predict(data, centroids):
    clusters = np.zeros((data.shape[0], 1))
    # For each instance in the dataset, define a distance vector of size k
    for index_x, x in enumerate(data):
        dist_vec = np.zeros((len(centroids), 1))
            
    # For each centroid, compute the distance between x and the centroid
        for index_centroid, centroid in enumerate(centroids):
            dist_vec[index_centroid] = euclidean(centroid, x)
            
            # Find the smallest distance, assign that distance to a cluster
            clusters[index_x, 0] = np.argmin(dist_vec)
    return clusters
```

## PCoA

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.collections import LineCollection

from sklearn.utils import shuffle
from sklearn import manifold
from sklearn.manifold import MDS
from sklearn.metrics import euclidean_distances
from sklearn.decomposition import PCA
```

```python
def PCoA(data, label, dissimilarity_mat='euclidean'):
    targets = list(range(0,10))
    for digit in targets:
        instances = [i for i in label if i==digit]
        print("Digit",digit,"appears ",len(instances), "times")

    if dissimilarity_mat == 'euclidean':
        transformer = MDS(n_components = 2, dissimilarity='euclidean', max_iter=10, n_init=1)
    else:
        transformer = MDS(n_components = 2, dissimilarity='precomputed', max_iter=10, n_init=1)

    X_transformed = transformer.fit_transform(data)
    plot_PCoA(X_transformed, label)
    
def plot_PCoA(*data):
    X,y=data
    mds=manifold.MDS(n_components=2)
    X_r=mds.fit_transform(X)

    ### graph
    fig=plt.figure()
    ax=fig.add_subplot(1,1,1)
    colors=((1,0,0),(0,1,0),(0,0,1),(0.5,0.5,0),(0,0.5,0.5),(0.5,0,0.5),
        (0.4,0.6,0),(0.6,0.4,0),(0,0.6,0.4),(0.5,0.3,0.2),)
    for label, color in zip( np.unique(y),colors):
        position=y==label
        ax.scatter(X_r[position,0],X_r[position,1],label="target= {0}".format(label))

    ax.set_xlabel("Y")
    ax.set_ylabel("X")
    ax.legend(loc="best")
    ax.set_title("PCoA for MNIST")
    plt.show()
    
```

```python
from sklearn.manifold import Isomap
```

```python
data = small_x
target = small_y

model = Isomap(n_components=2)
proj = model.fit_transform(data)
plt.scatter(proj[:, 0], proj[:, 1], c=target, cmap=plt.cm.get_cmap('jet', 10))
plt.colorbar(ticks=range(10))
plt.clim(-0.5, 9.5);
```

```python
from matplotlib import offsetbox

def plot_components(data, model, images=None, ax=None,
                    thumb_frac=0.05, cmap='gray'):
    ax = ax or plt.gca()
    
    proj = model.fit_transform(data)
    ax.plot(proj[:, 0], proj[:, 1], '.k')
    
    if images is not None:
        min_dist_2 = (thumb_frac * max(proj.max(0) - proj.min(0))) ** 2
        shown_images = np.array([2 * proj.max(0)])
        for i in range(data.shape[0]):
            dist = np.sum((proj[i] - shown_images) ** 2, 1)
            if np.min(dist) < min_dist_2:
                # don't show points that are too close
                continue
            shown_images = np.vstack([shown_images, proj[i]])
            imagebox = offsetbox.AnnotationBbox(
                offsetbox.OffsetImage(images[i], cmap=cmap),
                                      proj[i])
            ax.add_artist(imagebox)
```

```python
# Plot the 1s
data = small_x[small_y == 1]
fig, ax = plt.subplots(figsize=(10, 10))
model = Isomap(n_neighbors=10, n_components=2)
plot_components(data, model, images=data.reshape((-1, 28, 28)),
                ax=ax, thumb_frac=0.05, cmap='gray_r')
```

```python
# Plot the 2s
data = small_x[small_y == 2]
fig, ax = plt.subplots(figsize=(10, 10))
model = Isomap(n_neighbors=10, n_components=2)
plot_components(data, model, images=data.reshape((-1, 28, 28)),
                ax=ax, thumb_frac=0.05, cmap='gray_r')
```
